<h2 align="left"><font color=#000000>Problem Statement</font></h2>

The dataset consists of various attributes related to individuals' characteristics and lifestyle factors, collected to predict the risk of obesity. The objective is to develop a multiclass classification model to predict the obesity risk category of individuals based on their demographic information, dietary habits, physical activity levels, and other relevant factors.


<br> 
Here's a description of the columns for multiclass prediction of obesity risk in the form of a markdown table:

| **Column**                      | **Description**                                                                                        |
|--------------------------------|--------------------------------------------------------------------------------------------------------|
| **id**                         | Unique identifier for each individual                                                                 |
| **Gender**                     | Gender of the individual (e.g., Male, Female)                                                          |
| **Age**                        | Age of the individual in years                                                                         |
| **Height**                     | Height of the individual in centimeters                                                                |
| **Weight**                     | Weight of the individual in kilograms                                                                  |
| **family_history_with_overweight** | Indicates whether the individual has a family history of overweight (Yes or No)                     |
| **FAVC**                       | Indicates whether the individual frequently consumes high-calorie food (Yes or No)                  |
| **FCVC**                       | Frequency of consumption of vegetables (Numeric value)                                             |
| **NCP**                        | Number of main meals consumed per day (Numeric value)                                               |
| **CAEC**                       | Consumption of food between meals (Categorical: Frequently, Sometimes, Always, No)                  |
| **SMOKE**                      | Indicates whether the individual smokes (Yes or No)                                                 |
| **CH2O**                       | Daily water consumption in liters (Numeric value)                                                  |
| **SCC**                        | Calories consumed from snacks between meals (Yes or No)                                         |
| **FAF**                        | Physical activity frequency (Numeric value)                                                         |
| **TUE**                        | Time spent using technological devices (Numeric value)                                              |
| **CALC**                       | Consumption of alcohol (Categorical: Frequently, Sometimes, Always, No)                             |
| **MTRANS**                     | Mode of transportation used by the individual (Categorical: Automobile, Bike, Public_Transportation, Walking) |
| **NObeyesdad**                 | Obesity risk category of the individual (Multiclass: 7 different categories based on BMI)            |

```python
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
import numpy as np
from tqdm.notebook import tqdm


# Create a label encoder
label_encoder = LabelEncoder()

# Encode the target variable 'y' into integers
y_encoded = label_encoder.fit_transform(y)

# Specify the number of splits for Stratified K-Fold
splits = 5

# Initialize arrays to store training and validation accuracies
lst_train = []
lst_val = []

# Initialize array to store predicted probabilities for test set
predict_proba_test = np.empty((splits, X_test.shape[0], len(label_encoder.classes_)))

# Create Stratified K-Fold object
skf = StratifiedKFold(n_splits=splits, random_state=42, shuffle=True)

# Iterate through folds
for fold, (train_idx, val_idx) in tqdm(enumerate(skf.split(X, y_encoded))):
    # Define train set
    X_train = X.iloc[train_idx]
    y_train = y_encoded[train_idx]

    # Define validation set
    X_val = X.iloc[val_idx]
    y_val = y_encoded[val_idx]

    # Initialize XGBClassifier
    xgb_1 = XGBClassifier(
        learning_rate=0.1,
        n_estimators=2000,
        early_stopping_rounds=50,
        max_depth=4,
        min_child_weight=0.5,
        gamma=0.2,
        colsample_bytree=0.7,
        subsample=0.7,
        reg_alpha=1,
        objective='multi:softprob',
        nthread=-1,
        random_state=42
    )

    # Fit XGBClassifier on the training set
    xgb_1.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=False)

    # Calculate training and validation accuracies
    acc_train = accuracy_score(xgb_1.predict(X_train), y_train)
    acc_val = accuracy_score(xgb_1.predict(X_val), y_val)

    # Append accuracies to lists
    lst_train.append(acc_train)
    lst_val.append(acc_val)

    # Store predicted probabilities for the test set
    predict_proba_test[fold] = xgb_1.predict_proba(X_test)

# Print average training and validation accuracies
print(f"Model: xgb_1\nTrain Accuracy: {np.mean(lst_train):.4f}\nValidation Accuracy: {np.mean(lst_val):.4f}")

# Calculate average predicted probabilities for the test set
predictions_1 = predict_proba_test.sum(axis=0) / splits

# Decode the predicted classes
predicted_classes = label_encoder.inverse_transform(np.argmax(predictions_1, axis=1))
```

```python
# Create a dictionaries of list of models to evaluate performance with hyperparameters
models = { 
          'LinearRegression' : (LinearRegression(), {}),
          'SVR' : (SVR(), {'kernel': ['rbf', 'poly', 'sigmoid'], 'C': [1.0, 10.0, 100.0], 'epsilon': [0.01, 0.1, 1.0],}),
          'DecisionTreeClassifier' : (DecisionTreeClassifier(), {'max_depth': [None, 5, 10], 'criterion': ['gini', 'entropy'], 'random_state': [42]}),
          'AdaptiveBoostingClassifier' : (AdaBoostClassifier(), {'n_estimators': [10, 100], 'learning_rate': [0.1, 1.0], 'random_state': [42]}),
          'RandomForestClassifier' : (RandomForestClassifier(), {'n_estimators': [10, 100], 'max_depth': [None, 5, 10], 'random_state': [42]}),
          'KNeighborsClassifier' : (KNeighborsClassifier(), {'n_neighbors': np.arange(3, 100, 2), 'weights': ['uniform', 'distance']}),
          'GradientBoostingClassifier' : (GradientBoostingClassifier(), {'n_estimators': [10, 100], 'learning_rate': [0.1, 1.0], 'max_depth': [None, 5, 10], 'random_state': [42]}),
          'XGBClassifier' : (XGBClassifier(), {'n_estimators': [10, 100], 'learning_rate': [0.1, 1.0], 'max_depth': [None, 5, 10], 'random_state': [42]}),          
          
          }
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Create empty dictionaries to store evaluation metrics
mse_scores = {}
r2_scores = {}
mae_scores = {}

# train and predict each model with evaluation metrics as well making a for loop to iterate over the models
for name, (model, params) in models.items():
    # create a pipeline
    pipeline = GridSearchCV(model, params, cv=5)
    
    # fit the pipeline
    pipeline.fit(X_train, y_train)
    
    # make prediction from each model
    y_pred = pipeline.predict(X_test)
    
    # calculate evaluation metrics
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    
    # store evaluation metrics in dictionaries
    mse_scores[name] = mse
    r2_scores[name] = r2
    mae_scores[name] = mae
    
    # print the performing metric
    print(name, 'MSE: ', mse)
    print(name, 'R2: ', r2)
    print(name, 'MAE: ', mae)
    print('\n')

# Print the evaluation metrics
print("Mean Squared Error Scores:", mse_scores)
print("R2 Scores:", r2_scores)
print("Mean Absolute Error Scores:", mae_scores)
```
